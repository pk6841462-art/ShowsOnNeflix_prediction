```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['Year','Min_Age','IMDB_Rating']],df['Netflix'],test_size=0.3)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
prediction=model.predict([[2020,15,2.0]])
print(prediction)
```

```python
if prediction == 1:
    print("available on Netflix")
else:
    print("Not Available")
```

```python
probability=model.predict_proba([[2020,15,2.0]])
print(probability)
```

```python
accuracy=model.score(x_test,y_test)
print("Accuracy : ",accuracy)
```
